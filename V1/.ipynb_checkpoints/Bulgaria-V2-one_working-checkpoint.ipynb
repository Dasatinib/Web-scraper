```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

import requests
import pprint
import pyperclip
```

```python
html=urlopen("https://www.bda.bg/images/stories/documents/bdias/drugs2_list2_1.htm")
soup=BeautifulSoup(html)
```

```python
link_list=[]

for link in soup.find_all("a",href=re.compile("http://www.bda.bg")):
    temp=link.get("href")
    link_list.append(temp)
```

```python
open("output1.html","w+")
```

```python
urls = link_list
#scrape elements
for url in urls:
    response = requests.get(url)
    soup1 = BeautifulSoup(response.content)
    with open("output1.html", "a+", encoding='utf-8') as file:
        file.write(str(soup1))
```

```python
f = open("output1.html", encoding='UTF-8')
soup = BeautifulSoup(f, 'html.parser')
```

```python
open("table1.csv","w+")
```

```python
for drug in soup.find_all(True, {'class':["xl30","x131"]}):
    table=drug.get_text()
    with open("table1.csv", "a+", encoding='utf-8') as file:
        file.write(str(table))
```
